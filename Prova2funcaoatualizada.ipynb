```python
from dataclasses import dataclass
from scipy.integrate import solve_ivp
import numpy as np

@dataclass
class parametrosbioreator:
  """
  S = Concentração do substrato em mg/L
  Vmáx(T) = Velocidade máxima de consumo em mg/(L*h)
  Km = constante de Michaelis Menten em mg/L
  Ea = Energia de ativaçãowm J/mol
  T = Temperatura em K
  Tref = Temperatura de referência em K
  R = Constante universal dos gases em J/(mol*K)
  ti = tempo inicial em minutos
  tf = tempo final em minutos
  """
  Vmaxref: float = 5.0
  Ea: float = 30000.0
  Tref: float = 298.15
  R: float = 8.314
```

```python
def Vmaxt(T, param = parametrosbioreator):
  Vmaxt = param.Vmaxref*np.exp((-param.Ea/(param.R))*((1/param.T)-(1/param.Tref))) #definindo a função Vmáx
  return Vmaxt
```

```python
Km = 20

def dSdt(t, S, T, param):  #definindo a função da derivada
  S = S[0]
  dSdt = -Vmaxt(T, param)*(S/(Km + S))
  return dSdt
```

```python
def simula(t, S0, param, T ):   #simulando o processo do bioreator
  ti = 0
  tf = 30

  sol = solve_ivp(dSdt, t_span = (ti, tf), y0 = [S0], t_eval = t, args=(T, param))
  return sol.t, sol.y[0]
```
